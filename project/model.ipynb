```python
# ============================================
# 🌦️ Enhanced LSTM Model for Time Series Prediction
# ============================================
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

import tensorflow as tf
from tensorflow.keras.layers import (
    Input, Conv1D, BatchNormalization, Activation,
    Add, Bidirectional, LSTM, Dense, Dropout
)
from tensorflow.keras.models import Model
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
from tensorflow.keras.optimizers import RMSprop
from tensorflow.keras.regularizers import l2

# -----------------------------
# 1) Load & feature engineering
# -----------------------------
df = pd.read_csv("../src/ts_east/east_ppl.csv", parse_dates=["time"]).set_index("time")

# Gap handling
df = df.interpolate().fillna(method="bfill").fillna(method="ffill")

target_col = df.columns[2]                 
feature_cols = [c for c in df.columns if c != target_col]

# Light rolling stats on target (helps local shape)
df["roll_mean_5"] = df[target_col].rolling(5, min_periods=1).mean()
df["roll_std_5"]  = df[target_col].rolling(5, min_periods=1).std().fillna(0)

# Optional time features (kept small to avoid noise)
try:
    df["dayofweek"] = df.index.dayofweek
    df["month"]     = df.index.month
except Exception:
    pass

# ---------------------------------
# 2) Separate scaling (features/target)
# ---------------------------------
X_feats = df[[c for c in df.columns if c != target_col]].values
y_vals  = df[[target_col]].values

feat_scaler = StandardScaler()
tgt_scaler  = StandardScaler()

X_feats_scaled = feat_scaler.fit_transform(X_feats)
y_scaled       = tgt_scaler.fit_transform(y_vals)

# pack back together so target stays last for windowing
scaled_all = np.hstack([X_feats_scaled, y_scaled])

# -----------------------
# 3) Sliding window maker
# -----------------------
def make_dataset(data, n_steps=48):
    X, y = [], []
    for i in range(len(data) - n_steps):
        X.append(data[i:i+n_steps, :-1])  # features
        y.append(data[i+n_steps, -1])     # target (next step)
    return np.array(X), np.array(y)

N_STEPS = 48      # try 32 (more reactive) or 64/80 (smoother)
X, y = make_dataset(scaled_all, n_steps=N_STEPS)

split = int(len(X) * 0.8)
X_train, X_test = X[:split], X[split:]
y_train, y_test = y[:split], y[split:]

# ---------------------------------
# 4) Model: CNN blocks + residual + Bi-LSTM
# ---------------------------------
def conv_block(x, filters, k=2, wd=1e-4):
    """Conv -> BN -> ReLU"""
    x = Conv1D(filters, k, padding="same",
               kernel_regularizer=l2(wd), use_bias=False)(x)
    x = BatchNormalization()(x)
    return Activation("relu")(x)

inp = Input(shape=(X_train.shape[1], X_train.shape[2]))

# Initial projection (so residual dims match)
proj = Conv1D(64, 1, padding="same", use_bias=False)(inp)

# Conv block 1
x = conv_block(inp, 64, k=2)
x = conv_block(x, 64, k=2)

# Residual add (keeps high-freq info from input)
x = Add()([x, proj])

# Another lightweight conv to refine
x = conv_block(x, 64, k=2)

# Bidirectional LSTM for long context
x = Bidirectional(LSTM(64, return_sequences=False,
                       kernel_regularizer=l2(1e-4)))(x)

# Small MLP head
x = Dense(32, activation="relu")(x)
x = Dropout(0.08)(x)  # small dropout to retain detail
out = Dense(1)(x)     # linear output

model = Model(inp, out)
model.compile(optimizer=RMSprop(learning_rate=1e-3), loss="mae")
model.summary()

# ---------------------------------
# 5) Training controls
# ---------------------------------
early_stop   = EarlyStopping(monitor="val_loss", patience=14, restore_best_weights=True)
lr_scheduler = ReduceLROnPlateau(monitor="val_loss", factor=0.3, patience=3, min_lr=1e-5, verbose=1)

history = model.fit(
    X_train, y_train,
    epochs=150,
    batch_size=32,          # ↑ for smoother generalization
    validation_split=0.2,   # split within the train block
    callbacks=[early_stop, lr_scheduler],
    verbose=1
)

# ----------------
# 6) Evaluation
# ----------------
y_pred = model.predict(X_test)
y_pred_inv = tgt_scaler.inverse_transform(y_pred)
y_test_inv = tgt_scaler.inverse_transform(y_test.reshape(-1, 1))

mse = mean_squared_error(y_test_inv, y_pred_inv)
mae = mean_absolute_error(y_test_inv, y_pred_inv)
r2  = r2_score(y_test_inv, y_pred_inv)
print(f"\n📊 MSE: {mse:.4f}\n📉 MAE: {mae:.4f}\n🎯 R²: {r2:.4f}")

# ----------------
# 7) Plots
# ----------------
plt.figure(figsize=(7,4))
plt.plot(history.history["loss"], label="Train Loss")
plt.plot(history.history["val_loss"], label="Val Loss")
plt.title("Training & Validation Loss (MAE)")
plt.xlabel("Epochs"); plt.ylabel("MAE")
plt.legend(); plt.show()

# Align with real timestamps
test_index = df.index[-len(y_test_inv):]

plt.figure(figsize=(10,5))
plt.plot(test_index, y_test_inv, label="True")
plt.plot(test_index, y_pred_inv, label="Predicted")
plt.title("Predicted vs True (Inverse-Scaled)")
plt.xlabel("Time"); plt.ylabel(target_col)
plt.legend(); plt.show()

plt.figure(figsize=(10,4))
zoom = min(150, len(test_index))
plt.plot(test_index[:zoom], y_test_inv[:zoom], label="True")
plt.plot(test_index[:zoom], y_pred_inv[:zoom], label="Predicted")
plt.title("Zoomed (First 150 Test Points)")
plt.xlabel("Time"); plt.ylabel(target_col)
plt.legend(); plt.show()

# ----------------------------
# 8) One-step ahead forecast
# ----------------------------
future_scaled = model.predict(X[-1].reshape(1, N_STEPS, X.shape[2]), verbose=0)
future_value  = tgt_scaler.inverse_transform(future_scaled)[0,0]
print(f"\n🔮 Next-step forecast (original scale): {future_value:.4f}")
```
